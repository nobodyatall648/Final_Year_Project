# Merging 2 model predict result to Improve Accuracy

### Load models

```python
import os
import librosa
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model
from IPython import display as ipd
```

```python
dogBarkModel = 'dog_barks_classification.hdf5'
dogEmotionModel = 'dog_emotion_recognition.hdf5'

dogBarkLoadedmodel = load_model(os.path.join('model',dogBarkModel))
dogEmotionLoadedmodel = load_model(os.path.join('model',dogEmotionModel))
```

### Defining Functions

```python
def dogEmotionPred(filepath):
    img_rows = 224
    img_cols = 224
    
    img = cv2.imread(filepath)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
    img = cv2.resize(img,(img_rows,img_cols)).astype('float32')
    img /= 255
    img = np.expand_dims(img,axis=0)
    predict_label = dogEmotionLoadedmodel.predict(img)
    
    return predict_label
    
```

```python
def dogBarkPred(filepath):
    audio, sample_rate = librosa.load(filepath, res_type='kaiser_fast') 
    mfccs_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
    mfccs_scaled_features = np.mean(mfccs_features.T,axis=0) #scaling
    data = mfccs_scaled_features.reshape(1,-1)
    
    #predicted_label = chosen_model.predict_classes(data)
    predicted_label = dogBarkLoadedmodel.predict(data)
    
    return predicted_label
```

### Perform Weighted Averaging on 2 models prediction result 

To enhance the prediction result using dog emotion image & dog bark audio data

The dog emotion recognition model will have a higher accuracy comparing to dog bark recognition model. So the weights for dog emotion recognition model will be set higher comparing to dog bark recognition model to make the averaging accuracy higher depending on the dog emotion recognition model. <br>

dog emotion recognition model: 7 weight<br>
dog bark recognition model: 3 weight <br>

to select the best weightage, this will be future development.

```python
def average(a, b):
    combinedList = []
    combinedList.append(a.tolist())
    combinedList.append(b.tolist())
    
    dogPredArr = np.array(combinedList)
    predAvg = np.average(dogPredArr, axis=0, weights=[7,3])
    
    return predAvg
```

#### Showing prediction result before & after average

##### Explanation on output labels

dog emotion recognition = ['bow-wow', 'growling', 'howl'] <br>
dog bark recognition = ['happy', 'angry', 'sick']  <br>
<br>
bow-wow + happy => happy dog <br>
growling + angry => angry dog <br>
howl + sick => sick dog <br>

```python
labels = ['happy', 'angry', 'sick']

imgFile = os.path.join("test_data", "test_2.jpg")
barkFile = os.path.join("test_data", "test_2.wav")  

dogEmotionPredRsl = dogEmotionPred(imgFile)
dogBarkPredRsl = dogBarkPred(barkFile)

print("output meaning => [happy \t angry \t sick]", end="\n\n")
print("Before Average")
print("==============")
print("Dog Emotion Prediction: " + str([pred for pred in dogEmotionPredRsl[0]]))
print("Dog Bark Prediction: " + str([pred for pred in dogBarkPredRsl[0]]), end="\n\n")

dogPredAvg = average(dogEmotionPredRsl[0], dogBarkPredRsl[0])
predict_label = np.argmax(dogPredAvg, axis=0)

print("After Average")
print("=============")
print("Prediction (image + audio model): " + str([pred for pred in dogPredAvg]), end="\n\n")

print('[*] Dog is {}'.format(labels[predict_label]))
```

### Benchmarking accuracy before weighted average & after weighted average

#### Sample Size: 210 data (Each label 70 data)

Loading audio & images (into X & y)

```python
dogBarksAudioDir = "benchmark/audio"
labels = dict([('bow-wow',0),('growling',1),('howl',2)])
extracted_featuresAudio = []

#extracting audio features from each labelled directory with audio
for label in labels:
    fileDir = "%s/%s/" % (dogBarksAudioDir, label) 
    #getting wav files in directory
    files =  [f for f in os.listdir(fileDir) if ".wav" in f]
    
    for file in files:
        filePath = '%s%s' % (fileDir, file)      
        extracted_featuresAudio.append([filePath, labels[label]])
```

```python
current_path = os.getcwd()
data_path = os.path.join(current_path, 'benchmark/img')
data_folders = os.listdir(data_path)

img_rows = 224
img_cols = 224
extracted_featuresImg = []
labels = dict([('happy',0),('angry',1),('sick',2)])

for folder in data_folders:
    working_dir = os.path.join(data_path,folder)
    os.chdir(working_dir)
    current_list = os.listdir(os.getcwd())
    for img in current_list:
        extracted_featuresImg.append(['benchmark/img/'+folder+"/"+img, labels[folder]])        
os.chdir(current_path)
```

Sort y label 0 -> 2 acsending order

```python
extracted_featuresImg = sorted(extracted_featuresImg,key=lambda l:l[1])
```

```python
extracted_featuresAudio = sorted(extracted_featuresAudio,key=lambda l:l[1])
```

```python
def accCalc(correctAcc, sampleSize):
    acc = correctAcc / sampleSize
    
    return acc
```

Before Weighted Average Prediction Accuracy

```python
correctPred = 0

for i in range(len(extracted_featuresImg)):
    expectedLabel = extracted_featuresImg[i][1]
    predLabel = np.argmax(dogEmotionPred(extracted_featuresImg[i][0]), axis=1)[0]
    
    if(expectedLabel == predLabel):
        correctPred+=1
```

```python
dogEmotionAcc = accCalc(correctPred, len(extracted_featuresImg))
print("[*] Correct Predicted: " + str(correctPred))
print(dogEmotionAcc)
```

```python
correctPred = 0

for i in range(len(extracted_featuresAudio)):
    expectedLabel = extracted_featuresAudio[i][1]
    predLabel = np.argmax(dogBarkPred(extracted_featuresAudio[i][0]), axis=1)[0]
    
    if(expectedLabel == predLabel):
        correctPred+=1
```

```python
dogBarkAcc = accCalc(correctPred, len(extracted_featuresAudio))
print("[*] Correct Predicted: " + str(correctPred))
print(dogBarkAcc)
```

After Weighted Average (merging Img + Audio pred result)

```python
correctPred = 0

for i in range(len(extracted_featuresAudio)):
    expectedLabel = extracted_featuresAudio[i][1]
    
    barkPredRsl = dogBarkPred(extracted_featuresAudio[i][0])
    emotionPredRsl = dogEmotionPred(extracted_featuresImg[i][0])
    
    dogPredAvg = average(barkPredRsl[0], emotionPredRsl[0])
    dogPredLabel = np.argmax(dogPredAvg, axis=0)
    
    
    if(expectedLabel == dogPredLabel):
        correctPred+=1
```

```python
avgDogPredAcc = accCalc(correctPred, 210)
print("[*] Correct Predicted: " + str(correctPred))
print(avgDogPredAcc)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
label = ['dog emotion acc', 'dog bark acc', 'weighted average prediction acc']
colors = ['b','b','r']
accArr = [dogEmotionAcc, dogBarkAcc, avgDogPredAcc]

bar = ax.bar(label,accArr, color=colors)
plt.legend(bar[1:] , ['before weighted average','after weighted average'])
plt.ylabel('accuracy')
plt.ylim(ymin=min(accArr)-0.1, ymax=1.0)
plt.title('Accuracy Benchmark Before & After Merging 2 Model Prediction')
plt.show()
```
